## CLI Entry Point

The CLI is also deprecated but maintained for backward compatibility.

```python
#| export
def main():
    """DEPRECATED CLI entry point. Use nbdev_mcp.install CLI instead."""
    import argparse
    
    warnings.warn(
        "nbdev-mcp-install CLI is deprecated. Functionality is now in nbdev_mcp.install",
        DeprecationWarning,
        stacklevel=2
    )
    
    parser = argparse.ArgumentParser(
        prog="nbdev-mcp-install",
        description="DEPRECATED: Install nbdev-mcp configuration for AI tools."
    )
    
    subparsers = parser.add_subparsers(dest="command", help="Commands")
    
    # Status command
    subparsers.add_parser("status", help="Check installation status")
    
    # Install command
    install_parser = subparsers.add_parser("install", help="Install configuration")
    install_parser.add_argument(
        "tool", nargs="?",
        choices=["vscode", "claude", "codex", "ollama", "all"],
        default="all", help="Tool to configure (default: all)"
    )
    install_parser.add_argument("--python", "-p", help="Python interpreter path")
    install_parser.add_argument("--force", "-f", action="store_true", help="Overwrite existing")
    install_parser.add_argument("--dry-run", "-n", action="store_true", help="Show what would be done")
    
    args = parser.parse_args()
    
    if args.command == "status" or args.command is None:
        statuses = check_status()
        print(f"\nPython: {get_python_path()}\n")
        print("Installation Status:")
        print("-" * 60)
        for status in statuses:
            icon = "✓" if status['configured'] else "✗"
            exists = "exists" if status['exists'] else "missing"
            print(f"  {icon} {status['provider']:8} - {exists:8} - {status['path']}")
        print()
        
    elif args.command == "install":
        providers = None if args.tool == "all" else [args.tool]
        results = install_all(
            python_path=args.python,
            force=args.force,
            dry_run=args.dry_run,
            providers=providers
        )
        for result in results:
            icon = "✓" if result.success else "✗"
            print(f"{icon} {result.message}")
```
