```
#| export
from __future__ import annotations
import logging
import os
import sys
from typing import Optional, TextIO
```

## Logging

```
#| export
DEFAULT_LOG_FORMAT = "[%(asctime)s] %(levelname)s %(name)s: %(message)s"
"""Default log format string used by nbdev-mcp loggers."""

DEFAULT_LOG_NAME = "mcp.nbdev"
"""Default logger name for nbdev-mcp."""


def setup_logging(
    level: str | int | None = None,
    fmt: str | None = None,
    name: str = DEFAULT_LOG_NAME,
    stream: TextIO | None = None,
    force: bool = False,
) -> logging.Logger:
    """Configure and return the shared nbdev-mcp logger.
    
    Parameters
    ----------
    level : str | int | None
        Log level (e.g., ``'DEBUG'``, ``'INFO'``, ``logging.WARNING``).
        Defaults to ``NBMCP_LOG_LEVEL`` env var or ``'INFO'``.
    fmt : str | None
        Log format string. Defaults to :data:`DEFAULT_LOG_FORMAT`.
    name : str
        Logger name. Defaults to ``'mcp.nbdev'``.
    stream : TextIO | None
        Output stream for the handler (e.g., ``sys.stderr``).
        Defaults to ``sys.stderr``.
    force : bool
        If True, remove existing handlers before configuring.
        Useful for reconfiguring in notebooks.
    
    Returns
    -------
    logging.Logger
        Configured logger instance.
    """
    effective_level = level or os.environ.get("NBMCP_LOG_LEVEL", "INFO")
    effective_fmt = fmt or os.environ.get("NBMCP_LOG_FORMAT", DEFAULT_LOG_FORMAT)
    
    logger = logging.getLogger(name)
    
    if force:
        logger.handlers.clear()
    
    if not logger.handlers:
        handler = logging.StreamHandler(stream or sys.stderr)
        handler.setFormatter(logging.Formatter(effective_fmt))
        logger.addHandler(handler)
    
    logger.setLevel(effective_level)
    return logger


def set_log_level(level: str | int, name: str = DEFAULT_LOG_NAME) -> None:
    """Update a logger's level at runtime.
    
    Parameters
    ----------
    level : str | int
        New log level (e.g., ``'DEBUG'``, ``logging.INFO``).
    name : str
        Logger name to update. Defaults to ``'mcp.nbdev'``.
    """
    logging.getLogger(name).setLevel(level)


def set_log_format(fmt: str, name: str = DEFAULT_LOG_NAME) -> None:
    """Update a logger's format at runtime.
    
    Parameters
    ----------
    fmt : str
        New format string.
    name : str
        Logger name to update. Defaults to ``'mcp.nbdev'``.
    """
    logger = logging.getLogger(name)
    for handler in logger.handlers:
        handler.setFormatter(logging.Formatter(fmt))


def get_logger(name: str = DEFAULT_LOG_NAME) -> logging.Logger:
    """Get an existing logger by name without reconfiguring.
    
    Parameters
    ----------
    name : str
        Logger name. Defaults to ``'mcp.nbdev'``.
    
    Returns
    -------
    logging.Logger
        The logger instance.
    """
    return logging.getLogger(name)


# Initialize the default logger on import
log = setup_logging()
```
